```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("heart_disease.csv")
df.head()
```

```python
print(df.columns)
print(df.shape)
print(df.info())
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Blood Pressure'] = df['Blood Pressure'].fillna(df['Blood Pressure'].mean())
df['Cholesterol Level'] = df['Cholesterol Level'].fillna(df['Cholesterol Level'].mean())
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())

```

```python

df = df.dropna(subset=['Gender'])
df = df.dropna(subset=['Exercise Habits'])
df = df.dropna(subset=['Smoking'])
df = df.dropna(subset=['Family Heart Disease'])
df = df.dropna(subset=['Diabetes'])
df = df.dropna(subset=['High Blood Pressure'])
df = df.dropna(subset=['Low HDL Cholesterol'])
df = df.dropna(subset=['High LDL Cholesterol'])
df = df.dropna(subset=['Stress Level'])
df = df.dropna(subset=['Sleep Hours'])
df = df.dropna(subset=['Sugar Consumption'])
df = df.dropna(subset=['Triglyceride Level'])
df = df.dropna(subset=['Fasting Blood Sugar'])
df = df.dropna(subset=['CRP Level'])
df = df.dropna(subset=['Homocysteine Level'])
```

```python
df = df.fillna(df['Alcohol Consumption'].mode()[0])
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()

cols = [ 'Gender','Exercise Habits', 'Smoking', 'Family Heart Disease', 'Diabetes',
        'High Blood Pressure', 'Low HDL Cholesterol', 'High LDL Cholesterol',
       'Alcohol Consumption', 'Stress Level',
       'Sugar Consumption', 'Heart Disease Status']

for col in cols:
  df[col] = lb.fit_transform(df[col])

df.head()
```

```python
Cleaned_df = df.copy()
```

```python
# Standerd Scaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

for col in df.drop('Heart Disease Status',axis=1):
  Cleaned_df[col] = scaler.fit_transform(df[[col]])
Cleaned_df.head()
```

```python
X = Cleaned_df.drop("Heart Disease Status",axis=1)
Y = Cleaned_df['Heart Disease Status']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=45)
```

```python
# MODELS

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

# TEST

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,f1_score,recall_score
```

```python
models = {
    "Logistic Regression " : LogisticRegression(),
    "KNN" : KNeighborsClassifier(),
    "Dicision Tree" : DecisionTreeClassifier(),
    "Naive Bayes" : GaussianNB(),
    "SVM" : SVC()
}


```

```python


for name, model in models.items():
    model.fit(X_train,Y_train)
    Y_pred = model.predict(X_test)

    acc = accuracy_score(Y_test,Y_pred)
    recall = recall_score(Y_test,Y_pred)
    f1Score = f1_score(Y_test,Y_pred)
    classificationReport = classification_report(Y_test,Y_pred)

    print(f'------------------{name}------------------------')
    print(f'Accuracy -----------> {acc}')
    print(classificationReport)


```
