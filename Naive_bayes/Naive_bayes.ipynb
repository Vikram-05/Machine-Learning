```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df = df.drop(columns=['who','deck','alive'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()

cols = ['sex','embarked','class','adult_male','embark_town','alone']

for col in cols:
  df[col] = lb.fit_transform(df[col])

df.head()
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

newCols = [ 'pclass',  'age', 'sibsp', 'parch', 'fare','embarked', 'class',  'embark_town']

for col in newCols:
  df[col] = scaler.fit_transform(df[[col]])

df.head()
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df.isnull().sum()
```

```python
Cleaned_df = df.copy()
Cleaned_df.head()
```

```python
X = Cleaned_df.drop(columns=['survived'])
Y = Cleaned_df['survived']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=45)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
acc = accuracy_score(Y_test,Y_pred)
acc
```

```python
matrix = confusion_matrix(Y_test,Y_pred)
matrix
```

```python
report = classification_report(Y_test,Y_pred)
print(report)
```
