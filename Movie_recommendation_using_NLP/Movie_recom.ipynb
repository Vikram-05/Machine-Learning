```python
df['tags'] = df['tags'].apply(
    lambda text: " ".join([lemmatizer.lemmatize(word.lower(),pos='v') for word in word_tokenize(text) if word.lower() not in stopwords.words('english')])
)
```

```python
df['tags'][1]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)

vectors = tfidf.fit_transform(df['tags'])
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = df.reset_index(drop=True)
```

```python
df['title'] = df['title'].apply(
    lambda tit:  tit.lower()
)
```

```python
indices =pd.Series(df.index,index=df['title']).drop_duplicates()
indices
```

```python
def recommedation(title,max_size=10):
  title = title.lower()
  if title not in indices:
     return ["Movie not found"]
  idx = indices[title]
  sim_score = cosine_similarity(vectors[idx],vectors).flatten()
  similar_idx = sim_score.argsort()[::-1][1:max_size+1]
  return df['title'].iloc[similar_idx]
```
