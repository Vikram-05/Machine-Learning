```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df['sex'] = df['sex'].map({
    'male' : 0,
    'female' : 1
})
df.head()
```

```python
df.rename(columns={'sex':"isFemale"},inplace=True)
df.head()
```

```python
df = df.drop(columns= ['who','deck','alive'],axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['embarked'] = lb.fit_transform(df['embarked'])
df['class'] = lb.fit_transform(df['class'])
df['embark_town'] = lb.fit_transform(df['embark_town'])
df.head()
```

```python
# df['age'].isnull().sum()
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df['age'].isnull().sum()
```

```python
df['age'] = df['age'].astype(int)
```

```python
df.info()
```

```python
df['adult_male'] = df['adult_male'].astype(int)
df['alone'] = df['alone'].astype(int)
df.head()
```

```python
cleaned_df = df.copy()
```

```python
df.columns
```

```python
cols = [ 'pclass', 'age', 'sibsp', 'parch', 'fare',
       'embarked', 'class',  'embark_town']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df[cols] = scaler.fit_transform(df[cols])
df
```

```python
X = df.drop(columns=['survived'])
Y = df['survived']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.20,random_state=45)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)

model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
print(classification_report(Y_test,Y_pred))
```
